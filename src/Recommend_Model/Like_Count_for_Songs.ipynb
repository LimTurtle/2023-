## 좋아요 개수를 모든 곡에 분배한 json 데이터 생성
- 각 플레이리스트의 좋아요 개수를 포함된 모든 곡에 균등하게 분배
- 반영된 결과를 이후 사용하기 용이하도록 song_meta_with_likes.json 파일로 저장
- 실행시간이 매우 긴 일회성 코드이므로 Recommend_Model.ipynb 파일에서 별도로 분리함

```python
import json
import pandas as pd

with open('../Datasets/train.json', 'r', encoding='utf-8') as f:
    json_data = json.load(f)
    
train_data = pd.DataFrame(json_data)
train_data = train_data.drop(['id', 'plylst_title', 'updt_date'], axis=1)
train_data.head()
```

```python
with open('../Datasets/song_meta.json', 'r', encoding='utf-8') as f:
    json_data = json.load(f)
    
song_data = pd.DataFrame(json_data)
song_data.head()
```

```python
song_data['like_cnt_song'] = 0

for i in range(500, 1000):
    plist = train_data['songs'][i]
    likes = train_data['like_cnt'][i]
    for song in plist:
        song_data.loc[song_data['id'] == song, 'like_cnt_song'] += likes / len(plist)
        
sorted_song_data = song_data.sort_values(by='like_cnt_song', ascending=False)
sorted_song_data.head(10)
```

```python
song_data.to_json('../Datasets/song_meta_with_likes.json', orient='records')
```
